# DATA CLEANING

```python
data.isnull().sum()
```

```python

# Identify columns with missing values
columns_with_missing = ['Year', 'Duration', 'Genre', 'Rating', 'Votes', 'Director', 'Actor 1', 'Actor 2', 'Actor 3']

# Fill missing values with mean for numerical columns and mode for categorical columns
for column in columns_with_missing:
    if data[column].dtype == 'float64':
        data[column].fillna(data[column].mean(), inplace=True)
    else:
        data[column].fillna(data[column].mode()[0], inplace=True)

# Verify that missing values have been handled
data.isnull().sum()
```

# DATA TYPE CONVERSION

```python
data['Year'] = pd.to_numeric(data['Year'], errors='coerce')
data['Duration'] = pd.to_numeric(data['Duration'], errors='coerce')
data['Votes'] = pd.to_numeric(data['Votes'], errors='coerce')
print(data[['Year', 'Duration', 'Votes']].describe())
```

```python
# Convert the "Year" column to strings
data['Year'] = data['Year'].astype(str)

# Check for non-numeric values
non_numeric_years = data['Year'].loc[~data['Year'].str.isnumeric()]
print(non_numeric_years.unique())
```

```python
# Replace "nan" values with "0"
data['Year'] = data['Year'].replace('nan', '0')

# Convert the "Year" column to strings
data['Year'] = data['Year'].astype(str)

# Check for non-numeric values
non_numeric_years = data['Year'].loc[~data['Year'].str.isnumeric()]
print(non_numeric_years.unique())
```

```python
from wordcloud import WordCloud
# Create a function to generate word clouds
def generate_wordcloud(text_column, title):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(text_column.fillna('')))
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title)
    plt.axis('off')
    plt.show()

# Generate word clouds for Actor 1, Actor 2, Actor 3, and Director columns
generate_wordcloud(data['Actor 1'], 'Actor 1 Word Cloud')
generate_wordcloud(data['Actor 2'], 'Actor 2 Word Cloud')
generate_wordcloud(data['Actor 3'], 'Actor 3 Word Cloud')
generate_wordcloud(data['Director'], 'Director Word Cloud')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
# Define your target variable (what you want to predict) and features (columns used for prediction)
target_variable = 'Rating'  # Change to the actual column name for movie ratings
features = ['Year', 'Votes', 'Genre_Action', 'Genre_Comedy', 'Genre_Drama']  # Adjust the features as needed

# Split the data into training and testing sets (you can adjust the test_size and random_state)
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target_variable], test_size=0.2, random_state=42)

# Feature selection using SelectKBest and f_regression (adjust k as needed)
k_best = SelectKBest(score_func=f_regression, k='all')  # You can change 'all' to the number of top features you want
X_train = k_best.fit_transform(X_train, y_train)
X_test = k_best.transform(X_test)

# Standardize the features (mean=0, std=1) if needed
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Now, you have X_train, X_test (feature matrices) and y_train, y_test (target vectors) ready for modeling.



from sklearn.linear_model import LinearRegression  # You can choose a different model if needed
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the model (you can choose a different model here)
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Define your target variable (what you want to predict) and features (columns used for prediction)
target_variable = 'Rating'
features = ['Year', 'Votes', 'Genre_Action', 'Genre_Comedy', 'Genre_Drama']  # Adjust the features as needed

# Split the data into training and testing sets (you can adjust the test_size and random_state)
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target_variable], test_size=0.2, random_state=42)

# Initialize the Random Forest Regressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
rf_regressor.fit(X_train, y_train)

# Make predictions on the test data
y_pred_rf = rf_regressor.predict(X_test)

# Evaluate the model
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

# Print the evaluation metrics
print(f"Random Forest Regressor - Mean Squared Error: {mse_rf:.4f}")
print(f"Random Forest Regressor - R-squared: {r2_rf:.4f}")
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split

# Convert 'Year' column to numeric
data['Year'] = pd.to_numeric(data['Year'], errors='coerce')

# Define your target variable (what you want to predict) and features (columns used for prediction)
target_variable = 'Rating'  # Change to the actual column name for movie ratings
features = ['Year', 'Votes', 'Genre_Action', 'Genre_Comedy', 'Genre_Drama']  # Adjust the features as needed

# Split the data into training and testing sets (you can adjust the test_size and random_state)
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target_variable], test_size=0.2, random_state=42)

# Initialize the XGBoost Regressor
xgb_regressor = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, random_state=42)

# Train the model
xgb_regressor.fit(X_train, y_train)

# Make predictions on the test data
y_pred_xgb = xgb_regressor.predict(X_test)

# Calculate and print evaluation metrics (MSE and R-squared)
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
r2_xgb = r2_score(y_test, y_pred_xgb)

print(f"Mean Squared Error (XGBoost): {mse_xgb:.4f}")
print(f"R-squared (XGBoost): {r2_xgb:.4f}")
```

```python
 from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

# Define your target variable (what you want to predict) and features (columns used for prediction)
target_variable = 'Rating'  # Change to the actual column name for movie ratings
features = ['Year', 'Votes', 'Genre_Action', 'Genre_Comedy', 'Genre_Drama']  # Adjust the features as needed

# Preprocess the 'Year' column to remove parentheses and convert to integers
data['Year'] = data['Year'].astype(str).str.extract('(\d+)').astype(float)  # Convert to float first to handle NaN
data['Year'].fillna(0, inplace=True)
data['Year'] = data['Year'].astype(int)

# Convert the 'Votes' column to a string and then remove commas and other characters
data['Votes'] = data['Votes'].astype(str).str.replace(',', '', regex=True)  # Remove commas
data['Votes'] = data['Votes'].str.replace('M', 'e6', regex=True)  # Replace 'M' with 'e6' for million
data['Votes'] = data['Votes'].str.replace('k', 'e3', regex=True)  # Replace 'k' with 'e3' for thousand
data['Votes'] = pd.to_numeric(data['Votes'], errors='coerce').fillna(0).astype(int)  # Convert to integers

# Handle missing values in the target variable by imputing with the mean
data['Rating'].fillna(data['Rating'].mean(), inplace=True)

# Split the data into training and testing sets (you can adjust the test_size and random_state)
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target_variable], test_size=0.2, random_state=42)

# Initialize the SVM regressor
svm_regressor = SVR(kernel='linear')  # You can experiment with different kernels (e.g., 'linear', 'rbf', 'poly')

# Train the model
svm_regressor.fit(X_train, y_train)

# Make predictions on the test data
y_pred_svm = svm_regressor.predict(X_test)

# Evaluate the model
mse_svm = mean_squared_error(y_test, y_pred_svm)
r2_svm = r2_score(y_test, y_pred_svm)

# Print the evaluation metrics
print(f"Mean Squared Error (SVM): {mse_svm:.4f}")
print(f"R-squared (SVM): {r2_svm:.4f}")
```
